# Assignment

```python
attack_mean_hp = csvFile["HP"].mean()

def set_attack_hp(val):
    if val < attack_mean_hp:
        return "HP Low"
    elif val == attack_mean_hp:
        return "HP neutral"
    else:
        return "HP High"
```

```python
# "HP_high_low"  -> "HP High" , "HP Low"
csvFile["HP_high_low"] = csvFile["HP"].apply(set_attack_hp)
csvFile.head()
```

```python
attack_mean_speed = csvFile["Speed"].mean()

def set_attack_speed(val):
    if val < attack_mean_speed:
        return "Speed Low"
    elif val == attack_mean_speed:
        return "Speed neutral"
    else:
        return "Speed High"
```

```python
# "Speed_high_low"  -> "Speed High" , "Speed Low"
csvFile["Speed_high_low"] = csvFile["Speed"].apply(set_attack_speed)
csvFile.head()
```
